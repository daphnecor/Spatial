
The matrix containing the neural activity is the $T \times N$ matrix $\mathbf{X}$, where $T$ denotes the number of time slots and $N$ is the number of neurons. In this situation we assume that $T > N$. In standard analysis, we assume that $\mathbf{X}^\top\mathbf{X}$ is full rank, that is, $\text{rank}(\mathbf{X}^\top\mathbf{X}) = N$ which requires that $\text{rank}(\mathbf{X}) = N$. However, this assumption may not be valid in case of neural data.

As you can see, the principal components are derived from the covariance matrix $\mathbf{C}$. We assume that not all principal components are relevant, which means (implicitly) that we assume that some of the 'dimensions' in our neural activity matrix $\mathbf{X}$ are less relevant. That means that the variance of $\mathbf{X}$ can be approximated with $P < N$ dimensions (after making smart rotations). From an informal linear algebra perspective that means that the rank of $\mathbf{X}$ is approximately $P$.

You may now wonder how that relates to the neural manifold. We first observe that the realization $\mathbf{X}$ is generated by the intrinsic neural manifold and, assuming that $T$ and $N$ are sufficiently large, is representative for the intrinsic neural manifold. All information in $\mathbf{X}^\top\mathbf{X}$ is contained in $\mathbf{X}$ as $T > N$ and consequently $\text{rank}(\mathbf{X}) \geq \text{rank}(\mathbf{X}^\top\mathbf{X})$. If we assume that $\text{rank}(\mathbf{X}) = \text{rank}(\mathbf{X}^\top\mathbf{X})$ and we apply PCA to find the $P$ most relevant dimensions, we implicitly assume that the true underlying rank of $\mathbf{X}_{0}^\top\mathbf{X}_{0}$ (without noise) is equal to $P$, that is, $\text{rank}(\mathbf{X}_{0}^\top\mathbf{X}_{0}) = P$. Hence, we implicitly assume there exists an intrinsic (as $P < N$) neural manifold that generates the data.

## 1. Principal component analysis using the covariance method

```python
'''
Simulate some (correlated) data
'''

N = 1000
x = np.linspace(150, 190, num=1000) + np.random.normal(loc=0, scale=5, size=N)
y = x * 0.7 - 50 + np.random.normal(loc=0, scale=10, size=N)
```

---

> **2.** Take the eigendecomposition of the covariance matrix 

---

```python
# get the eigenvalues (D) and right eigenvectors (V)
D, V = np.linalg.eig(Cov)
```

---

> **3.** Sort the eigenvalues descending by their magnitude, and use this to sort their corresponding eigenvectors 

---

```python
# get the indices of the eigenvalues sorted in descending order
idxsort = np.argsort(D)[::-1]
# sort the eigenvalues and eigenvectors based on this order
D = D[idxsort]
V = V[:, idxsort]
```

---

> **4.** Convert eigenvalues to percentage change

---

```python
D = 100*D/np.sum(D) 
```

Let's look at the data and the two principal components
